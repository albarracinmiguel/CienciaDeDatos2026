## iloc — selección por posición

`iloc` (integer location) permite seleccionar filas y columnas **por su índice numérico**, independientemente del nombre.

La sintaxis es: `df.iloc[filas, columnas]`

- `df.iloc[0]` → primera fila
- `df.iloc[0:2]` → primeras dos filas (slice)
- `df.iloc[0, 1]` → fila 0, columna 1 (un solo valor)
- `df.iloc[:, 0]` → todas las filas, primera columna
- `df.iloc[[0, 2], 1:3]` → filas 0 y 2, columnas 1 a 2

```python
# Primera fila
print('Primera fila:')
print(df.iloc[0])
```

```python
# Primeras dos filas
df.iloc[0:2]
```

```python
# Filas 0 y 2, columnas 'edad' y 'nota' (posiciones 1 y 2)
df.iloc[[0, 2], 1:3]
```

## Dataset real: Titanic 🚢

El dataset del Titanic es un clásico de ciencia de datos. Contiene info de los pasajeros del Titanic: si sobrevivieron, clase del boleto, edad, sexo, tarifa pagada, etc.

Es ideal para practicar limpieza de datos, análisis exploratorio y modelos de clasificación.

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```python
titanic.info()
```

```python
# Tasa de supervivencia por sexo
titanic.groupby('Sex')['Survived'].mean().rename('tasa_supervivencia')
```

```python
# Edad promedio por clase
titanic.groupby('Pclass')['Age'].mean().rename('edad_promedio')
```

```python
# Cuántos valores nulos hay por columna
titanic.isnull().sum()
```
